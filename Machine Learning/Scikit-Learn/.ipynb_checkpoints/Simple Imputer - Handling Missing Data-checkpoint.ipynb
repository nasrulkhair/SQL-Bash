### Simple Imputer

- a scikit-learn clas which is helpful in handling the missing data in the predictive model dataset
- It replaces the NaN values with a specified placeholder.


```python
# Importing libraries

import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.compose import make_column_transformer
```

```python
# Example 1: Using mean as the stratgey var

data = {"farthest_run_kilo": [50,62, np.nan, 100,26,13,31,50], "runner": [1, 2, 3, 4, 5, 6, 7, 8]}
df = pd.DataFrame(data)
df

imp_mean = SimpleImputer(strategy = "mean")  # can use median / mode
df_imputed = pd.DataFrame(imp_mean.fit_transform(df), columns = df.columns)
df_imputed
```

```python
#Example 2: Use constant value

data2 = {"farthest_run_kilo": [50,62, 12, 100,26,13,31,50], "runner": ["Alice", "Bob", np.nan, "Diana", "Eve", "Frank", "Grace", "Hannah"]}
df_const = pd.DataFrame(data2)

imp_const = SimpleImputer(strategy = "constant", fill_value = "Nas")

df_imputed_const = pd.DataFrame(imp_const.fit_transform(df_const), columns = df_const.columns)
df_imputed_const
```

```python
# Example 3: multiple columns imputing

data_3 = {"farthest_run_kilo": [50,62, np.nan, 100,26,13,31,50], "runner": ["Alice", "Bob", np.nan, "Diana", "Eve", "Frank", "Grace", "Hannah"]}
df3 = pd.DataFrame(data_3)

# make column Transform
ct = make_column_transformer(
    (imp_const, ["runner"]),
    (imp_mean, ["farthest_run_kilo"]),
    remainder = "drop"
)

ct.set_output(transform = 'pandas')
```

```python
df_pandas = ct.fit_transform(df3)
df_pandas
```
