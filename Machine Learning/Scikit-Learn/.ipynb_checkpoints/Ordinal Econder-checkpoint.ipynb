### Ordinal Encoder

- Ordinal Encoding is a preprocessing technique used for converting categorical data into numeric values that preserce their inherent ordering
- It is useful when working with machine learning models like neural networks taht expect numerical input features
- Benefits :
  1. Encoding categorical data into numeric forms that algorithms can understand
  2. Retaining the ordinal information between categories that is lost with one-hot encoding

```python
# Importing Libraries

import numpy as np
import pandas as pd
# from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
```

```python
# Reading the file

d = {"sales": [100000, 222000, 1000000, 522000, 111111, 222222, 1111111, 20000, 75000, 90000, 1000000, 10000],
    "city": ["Tampa", "Tampa", "Orlando", "Jacksonville", "Miami", "Jacksonville", "Miami", "Miami", "Orlando", "Orlando", "Tampa", "Tampa"],
    "size": ["Small", "Medium", "Large", "Large", "Small", "Medium", "Large", "Small", "Medium", "Medium", "Medium", "Small"]}

df = pd.DataFrame(d)
df.head()
```

```python
# city - nominal data / size - ordinal data

# specifying the order
df["size"].unique()
sizes = ['Small', 'Medium', 'Large']
```

```python
ord_enc = OrdinalEncoder(categories = [sizes])
ord_enc
```

```python
df ["sizes"] = ord_enc.fit_transform(df[["size"]])
df.head()
```

```python
# making sure the comparison by sorting the data using 'sizes' column
df.sort_values(by = "sizes", ascending = True)

# small - 0 / medium - 1 / large - 2
```

```python
# remove unwanted column

df = df.drop(columns = "size")
df.head(10)
```
