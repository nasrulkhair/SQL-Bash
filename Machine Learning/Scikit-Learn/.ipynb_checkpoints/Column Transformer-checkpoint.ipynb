### Columns Tranformer

- Allows you to selectively apply data preparations transforms to different columns in a dataset
- paryicular useful when the dataset have amix of numerical and categorical data that require different preprocesing steps
- Why???
  1. Selective Transformation: Apply specific transformations to subsets of columns.
  2. Pipeline Integration: Easily integrate with scikit-learn's Pipeline for streamlined workflows.
  3. Code Organization: Encapsulate preprocessing logic in a single, maintainable object.

```python
# Importing Libraries

import pandas as pd

d = {"sales": [100000, 222000, 1000000, 522000, 111111, 222222, 1111111, 20000, 75000, 90000, 1000000, 10000],
    "city": ["Tampa", "Tampa", "Orlando", "Jacksonville", "Miami", "Jacksonville", "Miami", "Miami", "Orlando", "Orlando", "Tampa", "Tampa"],
    "size": ["Small", "Medium", "Large", "Large", "Small", "Medium", "Large", "Small", "Medium", "Medium", "Medium", "Small"]}

df = pd.DataFrame(d)
df
```

```python
# Example 1: multiple columns encoding (remainder = passthrough)

# library for ohe and ode

from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder

ohe = OneHotEncoder(sparse_output = False)  # to get outpu as pandas dataframe
ode = OrdinalEncoder(categories=[["Small", "Medium", "Large"]])
```

```python
# library for column transformation

from sklearn.compose import make_column_transformer

ct  = make_column_transformer(
    (ohe, ["city"]),
    (ode, ["size"]),
    remainder = "passthrough"
)

ct.set_output(transform = "pandas")
```

```python
df_pandas = ct.fit_transform(df)
df_pandas
```

```python
# Example 2: multiple columns encoding (remainder = "drop")

ct2  = make_column_transformer(
    (ohe, ["city"]),
    (ode, ["size"]),
    remainder = "drop"
)

ct2.set_output(transform = "pandas")
```

```python
df_drop = ct2.fit_transform(df)
df_drop
```

```python
# Example 2: multiple columns encoding (remainder = "drop" & "passthrough")

ct3  = make_column_transformer(
    (ohe, ["city"]),
    ("passthrough", ["size"]),
    remainder = "drop"
)

ct3.set_output(transform = "pandas")
```

```python
df_pass_drop = ct3.fit_transform(df)
df_pass_drop
```
