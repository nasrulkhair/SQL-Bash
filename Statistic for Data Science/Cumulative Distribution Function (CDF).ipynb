Cumulative Distribution Function (CDF)

- Probability of random variable X will take avalue less than or equeal to a particular value X.
- Range: 0 - 1
- Formulae: F(x) = P(X <= x)
  

What you can calculate?
1. AtX or Below
2. Range of Values ( Example: -1 to 1 / Upper Bound - Lower Bound )
3. At X or Higher (1- CDF)

```python
# importing libraries to find the CDF
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Example 1: Manual Calculation

data_x = [2, 3, 3, 5, 7]
sorted_data = np.sort(data_x)
data_len = len(sorted_data)

cdf_values = []

for i in range(data_len):
    cdf_value = np.sum(sorted_data <= sorted_data[i]) / data_len
    cdf_values.append(cdf_value)

cdf_values
```

```python
# Example 2: CDF at a single point


# Creating the random sample
np.random.seed(12)

mean = 0
std = 1
size = 1000


data = np.random.normal(loc = mean, scale = std, size = size)

cdf_neg_1 = norm.cdf(-1, loc = data.mean(), scale = data.std())
cdf_neg_1_percent = round((cdf_neg_1 * 100),2)
print(f" total of {cdf_neg_1_percent}% values are <= than -1")
```

```python
# Example 3: CDF Range (-2, 2)

upper_cdf = norm.cdf(2, loc = data.mean(), scale = data.std())
lower_cdf = norm.cdf(-2, loc = data.mean(), scale = data.std())

range_cdf = upper_cdf - lower_cdf 
range_cdf_percent = round((range_cdf * 100), 2)
print(f" Total {range_cdf_percent}% falls in between of -2 and 2")
```

```python
# Example 4: CDF roght value greater than 2

value_greater_than_2 = 1 - norm.cdf(2, loc = data.mean(), scale = data.std())
value_greater_than_2_percent = round((value_greater_than_2 * 100), 2)
print(f" Total {value_greater_than_2_percent}% of the values are greater than 2")
```

```python
# Example 5: Graph

sns.ecdfplot(data, label = "CDF")
plt.title("CDF Normal Distribution")
plt.xlabel("Data Values")
plt.ylabel("Cumulative Probability Distribution")
plt.legend()
plt.grid(True)
plt.show()
```
