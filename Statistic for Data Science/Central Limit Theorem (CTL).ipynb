## Central Limit Theorem

- The central limit theorem states that if we take many samples from any population distribution, the distribution of the sample means will approach a normal distribution.
- Things to do:
  1. Start with a Population
  2. Take Samples fromthe distribution (more than 25 numbers for a sample)
  3. Find the Mean for each sample
  4. Plot the Means on a histogram
  5. More Samples = Closer to a Normal Distribution

```python
# Importing required Libraries

import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
# Example 2: Binomial Distribution

# setting the seed
np.random.seed(13)

n_trials = 10
p_success = 0.5
population_size = 100000

binomial_data = np.random.binomial(n_trials, p_success, population_size)

sns.histplot(binomial_data, kde = False)
plt.title("Binomial Data")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
sample_size_bino = 200
n_sample_bino = 1000
bino_mean = []

for i in range(n_sample_bino):
    sample = np.random.choice(binomial_data, size = sample_size_bino, replace = False)
    sample_mean_bino = sum(sample) / sample_size_bino
    bino_mean.append(sample_mean_bino)

plt.hist(bino_mean, bins = 200, edgecolor = "black", density = False)
plt.title("Binomial Means")
plt.xlabel("Frequency")
plt.ylabel("Sample Means")
```

```python
# Example 3: Poisson Distribution
# setting the seed
np.random.seed(13)

lambda_poisson = 3
population_size = 100000
poisson_data = np.random.poisson(lambda_poisson, population_size)

sns.histplot(poisson_data, kde = False)
plt.title("Poisson Data")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
sample_size_pois = 30
num_samples_pois = 1000
poiss_mean = []

for i in range(num_samples_pois):
    sample = np.random.choice(poisson_data, size = sample_size_pois, replace = False)
    sample_mean_pois = sum(sample) / sample_size_pois
    poiss_mean.append(sample_mean_pois)

plt.hist(poiss_mean, bins = 30, edgecolor = "black", density = False)
plt.title("Poisson Means")
plt.xlabel("Frequency")
plt.ylabel("Sample Means")
```
