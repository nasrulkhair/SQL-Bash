## PPF

- Also known as the quantile function or inverse cumulative distribution function Opposite of the CDF
- Input is a decimal probability ( 0 to 1 )
- Output : Value
- Formulae :
  x = (F**-1)(p)
  


What You can calculate

1. At % or Below:
   
    a. Just Use PPF with decimal
3. Range of values:
   
    a. Lower %: Upper%
5. At X or Higher:

    a. 1-decimal
   
        i. Example top 25% = PPF(0.75)

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
# Create our normal distribution
np.random.seed(17)

# set up the parameter 
mean = 0
std_dev =  1
size = 1000

# Generating the data
data = np.random.normal(loc = mean, scale = std_dev, size = size)
```

```python
# Example 1: lowest 20% and 70% of the value

# Lowest 20%
ppf_20 = norm.ppf(0.2, loc = data.mean(), scale = data.std())
print(ppf_20)
# lowest 70%
ppf_70 = norm.ppf(0.7, loc = data.mean(), scale = data.std())
print(ppf_70)
```

```python
# Example 2: Find the data points between 25% and 50%
```

```python
ppf_25 = norm.ppf(0.25, loc = data.mean(), scale = data.std())
print(ppf_25)
ppf_50 = norm.ppf(0.5, loc = data.mean(), scale = data.std())
print(ppf_50)
```

```python
# Example 3: Find the top 20% of values and top 100%
ppf_top_20 = round(norm.ppf((1-0.2), loc = data.mean(), scale = data.std()),4)
print(ppf_top_20)

ppf_100 = norm.ppf(1, loc = data.mean(), scale =data.std())
print(ppf_100)
```

```python
#Example 4: Numpy Percentile

ppf_20_np = np.percentile(data, 20)
print(ppf_20_np)
ppf_70_np = np.percentile(data, 70)
print(ppf_70_np)
```

```python
# Example 5: Visualize the results

sorted_data = np.sort(data)
cumulative_probabilities = np.linspace(0, 1, len(sorted_data), endpoint = False)

plt.plot(cumulative_probabilities, sorted_data, marker = "o", linestyle = "none", label = "PPF")
plt.title("PPF of Normally Distributed Data")
plt.xlabel("Cumulative Probability")
plt.ylabel("Data Values")
plt.legend()
plt.grid(True)
plt.show()
```
