# Poisson Distribution

- The Poisson Distribution is a probability that describes the number of events occuring within a fixed interval of time or space.
- Examples:
  1. Number of subscribers joining a Youtube channel in a day.
  2. Website Visitors per hour
  3. Chargebacks per Quarter

- It deals with counts of events that are non-negative integer values (0, 1, 2...)
- Terms to know:
  1. Lambda (The average rate (mean) at which events occir in given in a given interval)
  2. K (Number events you are calculating the probability for)
     - Example: Test 50 subscribers in a day OR 25 subscribers in a day
  

**Poisson Distribution - Concepts**

- PMF (Probbaility Mass Function) - Exact probability of a Number of Events happening
- CDF (Cumulative Distribution Function) - Probability Value less or equal to Number of Events.

```python
# Importing Libraries

import numpy as np
from scipy.stats import poisson
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Example 1: Generate random poisson from numpy
np.random.seed(12)

lambda_value = 5
size = 10
poisson_sample = np.random.poisson(lam = lambda_value, size =size)
poisson_sample
```

```python
# Example 2: Generate random poisson from scipy

poisson_sample_2 = poisson.rvs(mu = lambda_value, size = size, random_state = 11)
poisson_2_mean = poisson_sample_2.mean()
poisson_2_std = poisson_sample_2.std()
```

```python
# Example 3: Posson PMF Single Values

lambda_value = 5
k = 3

prob = poisson.pmf(k, lambda_value)
prob
```

```python
# Example 4: PMF multiple values

prob2 = poisson.pmf(poisson_sample, lambda_value)
prob2
```

```python
# Example 5: CDF at a single point

prob_cdf = poisson.cdf(k, lambda_value)
prob_cdf
```

```python
# Example 6: CDF at a multiple point

prob_cdf = poisson.cdf(poisson_sample, lambda_value)
prob_cdf
```

```python
# Example 7: Stem Plot - PMF

lambda_value = 5
k_values = np.arange(0, 15)


pmf_values = poisson.pmf(k_values, lambda_value)
pmf_values

# Visualization

plt.stem(k_values, pmf_values)
plt.title("Poisson Distribution")
plt.xlabel("Number of Events")
plt.ylabel("Probability")
plt.show()
```

```python
# Example 7: Histogram

sample_size = 1000
data = np.random.poisson(lambda_value, sample_size)

plt.hist(data, bins = np.arange(0,16)-0.5, edgecolor="black", alpha = 0.7)
plt.title("Poisson Distribution")
plt.xlabel("Number of Events")
plt.ylabel("Count")
plt.show()
```
