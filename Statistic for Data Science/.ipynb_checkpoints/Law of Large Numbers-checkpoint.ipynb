Law of Large Numbers

- The Law of Large Numbers shows that while individual outcomes are uncertain, the average of a large number of outcomes tends to be predictable to the expected value (convergence)

  1. Perform experiment many times
  2. The experiments averages will reach the true average as increase number of trials
  3. Observed probability goes to the theoretical probability

Examples:
 - Flipping a coin (Bernoulli Distribution)
 - Data Science Salary in a region (Normal Distribution)

```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Example 1: Bernoulli

np.random.seed(15)
n_trials = 10000

# create data for coin flips between 0 - 1
coin_flips = np.random.randint(0, 2, size = n_trials)

# finding the cumulative mean for both flip side
cumulative_mean = np.cumsum(coin_flips) / np.arange(1, n_trials + 1)
cumulative_mean
```

```python
# Plotting the graph

plt.figure(figsize=(10, 5))
plt.plot(cumulative_mean, label = "Cumulative Mean")
plt.axhline(y=0.5, color = "r", linestyle = "--", label = "Expected Value 0.5")
plt.xlabel("Number of Trials")
plt.ylabel("Cumulative Mean")
plt.title("Law of Large Numbers Coin Flips")
plt.legend()
plt.show()
```

```python
# Example 2: Data Science Salaries (Normal Distribution)

n_samples = 10000

mu = 120000
sigma = 30000
salaries = np.random.normal(loc = mu, scale = sigma, size = n_samples)
salaries
```

```python
cumulative_mean_salaries = np.cumsum(salaries) / np.arange(1, n_samples + 1)
cumulative_mean_salaries
```

```python
# Plotting the cumulative mean salaries

plt.figure(figsize=(10, 5))
plt.plot(cumulative_mean_salaries, label = "Cumulative Mean")
plt.axhline(y=mu, color = "r", linestyle = "--", label = "Expected Salary $120,000")
plt.xlabel("Number of Samples")
plt.ylabel("Cumulative Mean")
plt.title("Law of Large Numbers Normal Distribution DS Salaries")
plt.legend()
plt.show()
```
