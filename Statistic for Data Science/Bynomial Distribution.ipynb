## Binomial Distribution

The binomial distribution represents the probability distribution variable that counts the number of successes in a fixed number of independent trials. Each trial results in either success (with probability p) or failure (with probability 1-p)

Examples:

    - Coin Flip (Heads/Tails)
    - New Subscriber from watchin a YouTube Video (Subscribe / Not Subcribe)
    - Pitcher throwing astrike (Strike / Ball)

Discrete

Terms to know:
- success (p) (also known as probability at one trial)
    - 2.5% of views turn into subscribers
         - P(X=1) = 0.025
         - P(X = 0) 0.975
    - Trials (n)
        - Video Gets 1,000 Views
    - Size (Number of Trials)
        Look at 100 different videos

### Bernoulli Distribution - Concepts

- Binomial Outcomes of multiple trials of an event, bernoulli one trial of an event.
- PMF (Probability Mass Function) - Exact Probability of a Number of Events happening
- CDF (Cumulative distribution function) - Probability Values less or equal to Number of Events

```python
# Importing Libraries

import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Example 1: Create Binomial in numpy - Youtube Subscription

# Creating the random data
np. random.seed(12)

p_success_subs = 0.025
n_trials_subs = 1000     
size_videos = 100        # every 100 videos, 1000 views, probability of 2.5% become subscribers

numpy_samples = np.random.binomial(n = n_trials_subs, size = size_videos, p = p_success_subs)
numpy_samples
```

```python
# Example 2: Coin Flip using Scipy

np.random.seed(12)

p_success_coin = 0.5
n_trials_coin =  50
size_coin_flips = 25

scipy_sample = binom.rvs(n = n_trials_coin, p = p_success_coin, size = size_coin_flips)
scipy_sample
```

```python
# Example 3: PMF
# probability of exact occurences

prob_exact_occurences = round((binom.pmf(n = n_trials_coin, k = 25, p = p_success_coin) *100), 2)    # k - occurences of happening
print(f" The probability of hitting the head 25 times : {prob_exact_occurences}%")
```

```python
# Example 4: CDF
# less or equal to K occurences

binom_cdf = round((binom.cdf(n = n_trials_subs, k = 25, p = p_success_subs) * 100), 2)
print(f" {binom_cdf}% probability that we hit 25 or less number of subscribers on a video that has 1000 views")
```

```python
# Example 5: Coin Histogram

plt.figure(figsize = (8,6))
plt.hist(scipy_sample, bins = 10, color = "skyblue", edgecolor = "black")
plt.title("Histogram coin flips")
plt.xlabel("Success")
plt.ylabel("Frequency")
plt.show()
```

```python
# Example 6: Plot coin flip PMF
x = np.arange(0, n_trials_coin + 1)
pmf = binom.pmf(x, n = n_trials_coin, p = p_success_coin)

plt.figure(figsize = (8,6))
plt.stem(x, pmf)
plt.title("Histogram coin flips")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.show()
```

```python
# Example 7: CDF

x = np.arange(0, n_trials_coin + 1)
cdf = binom.cdf(x, n = n_trials_coin, p = p_success_coin)

plt.figure(figsize = (8,6))
plt.plot(x, cdf, "b-", lw = 2)
plt.title("CDF Coin flips")
plt.xlabel("Success")
plt.ylabel("Cumulative Probability")
plt.show()
```
