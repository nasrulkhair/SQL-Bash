## Skewness - Right, Left & Symmetric Distribution

- A skewed distribution is one where the data is not symetrically distributed around the mean. There are two types of skewed distributions:
    1. Right and Left Skewed: positive and negative skewed
    2. Non Skewed: Symmetric (Normal Distribution)
 

**Right Skewed (Positively Skewed)**

- The right tail (larger values) is longer, meaning the mass of the distribution is concentrated on the left. In such distributions, the mean is typically greater that the median.

**Left Skewed (Negatively Skewed)**

- The left tail (smaller values) is longer, emaning the mass of the distribution is concentrated on the righ. in such distributions, the mean is typically less that the median.

**Symmetric (Normal) Distribution**

- A symmetric distribution is one where the left and right sides of the distribution are mirror images of each other. As a result, the mean, median and mode are all equal andlocated at the center of the distribution.

```python
# imorting libraries

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Creating data

np.random.seed(42)

right_data = np.random.exponential(scale = 2, size = 1000)
left_data = np.random.exponential(scale = 2, size = 1000) * -1 + 7
normal_data = np.random.normal(loc = 5, scale = 1.5, size = 1000)

```

```python
# Example 1: Mean, median & mode

# Mean
mean_right = round(np.mean(right_data), 2)
mean_left = round(np.mean(left_data), 2)
mean_norm = round(np.mean(normal_data), 2)

# Median
median_right = round(np.median(right_data), 2)
median_left = round(np.median(left_data), 2)
median_norm = round(np.median(left_data), 2)

# Mode
mode_right = round(stats.mode(right_data, keepdims=True)[0][0], 2)
mode_left = round(stats.mode(left_data, keepdims = True)[0][0], 2)
mode_norm = round(stats.mode(normal_data, keepdims = True)[0][0], 2)
```

```python
# Example 2: Plotting the distributions


plt.figure(figsize = (21,6))


# Subplot right_skewed
plt.subplot(1, 3, 1)
sns.histplot(right_data, kde = True, color = "skyblue")
plt.axvline(mean_right, color = "r", linestyle = "--", label = f"mean: {mean_right}")
plt.axvline(median_right, color = "blue", linestyle = "--", label = f"median: {median_right}")
plt.axvline(mode_right, color = "black", linestyle = "--", label = f"mode: {mode_right}")
plt.title("Right Skewed Distribution")
plt.legend()

# Subplot left_skewed
plt.subplot(1, 3, 2)
sns.histplot(left_data, kde = True, color = "green")
plt.axvline(mean_left, color = "r", linestyle = "--", label = f"mean: {mean_left}")
plt.axvline(median_left, color = "blue", linestyle = "--", label = f"median: {median_left}")
plt.axvline(mode_left, color = "black", linestyle = "--", label = f"mode: {mode_left}")
plt.title("Left Skewed Distribution")
plt.legend()

# Subplot normal_distribution
plt.subplot(1, 3, 3)
sns.histplot(normal_data, kde = True, color = "yellow")
plt.axvline(mean_norm, color = "r", linestyle = "--", label = f"mean: {mean_norm}")
plt.axvline(median_norm, color = "blue", linestyle = "--", label = f"median: {median_norm}")
plt.axvline(mode_norm, color = "black", linestyle = "--", label = f"mode: {mode_norm}")
plt.title("Normal Distribution")
plt.legend()
```

```python
# Example 3: Ploting the Box Plot (Whisker)

plt.figure(figsize = (15,5))

plt.subplot(1, 3, 1)
sns.boxplot(y = right_data, color = "skyblue")
plt.title("Right Skewed Distribution", fontsize=12)
plt.ylabel("Values", fontsize = 10)


plt.subplot(1, 3, 2)
sns.boxplot(y = left_data, color = "green")
plt.title("Left Skewed Distribution", fontsize=12)
plt.ylabel("Values", fontsize = 10)


plt.subplot(1, 3, 3)
sns.boxplot(y = normal_data, color = "yellow")
plt.title("Normal Distribution", fontsize = 12)
plt.ylabel("Values", fontsize = 10)


# Adjusting layout for better spacing
plt.tight_layout()

# Increase spacing between subplots
plt.subplots_adjust(wspace = 0.6, top = 0.8)

plt.show()
```
