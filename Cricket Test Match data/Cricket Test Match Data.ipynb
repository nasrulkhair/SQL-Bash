# Calculations

```python
#1. What is the Average Career Lenghth

df["Career_Length"].mean()
```

```python
#2. AVG Batting Strike Rate for Cricketers Who played over 10 years
df[df["Career_Length"] >10]["Battling_Strike_Rate"].mean()
```

```python
#3. Find cricketers who played before 1960
```

```python
df[df["Rookie_Year"] < 1960]["Player_Name"].count()
```

```python
#4. Max Highest Inns Score by Country

df.groupby("Country")["Highest_Inns_Score"].max().to_frame("HihghInnCountry").reset_index().sort_values("HihghInnCountry",ascending = False)
```

```python
#5. Hundreds, Fifties, ducks (0) AVG by Country
df.groupby("Country")[["100","50","0"]].mean()
```
